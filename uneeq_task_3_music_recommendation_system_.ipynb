```python
df['text'] = df['text'].str.lower().replace(r'^\w\s', ' ').replace(r'\n', ' ', regex = True)
```

```python
!pip install nltk
import nltk
nltk.download('punkt')

import nltk
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

def tokenization(txt):
    tokens = nltk.word_tokenize(txt)
    stemming = [stemmer.stem(w) for w in tokens]
    return " ".join(stemming)

df['text'] = df['text'].apply(lambda x: tokenization(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Combine text features into a single column
df['combined_features'] = df['artist'] + ' ' + df['song'] + ' ' + df['text']

# Initialize the TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(max_features=50000)

# Fit and transform the combined text data
tfidf_matrix = tfidf_vectorizer.fit_transform(df['combined_features'])

# Convert TF-IDF matrix to a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), index=df.index)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute the cosine similarity matrix
cosine_sim_matrix = cosine_similarity(tfidf_matrix)

def recommend_songs(song_title, cosine_sim_matrix, df, top_n=10):
    # Get the index of the song that matches the title
    song_idx = df[df['song'].str.contains(song_title, case=False)].index[0]

    # Get the pairwise similarity scores of all songs with that song
    sim_scores = list(enumerate(cosine_sim_matrix[song_idx]))

    # Sort the songs based on similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the indices of the top_n most similar songs
    sim_scores = sim_scores[1:top_n+1]

    # Get the titles of the most similar songs
    song_indices = [i[0] for i in sim_scores]

    return df[['artist', 'song']].iloc[song_indices]
```

```python
recommendations = recommend_songs("she", cosine_sim_matrix, df, top_n=10)
print(recommendations)
```
